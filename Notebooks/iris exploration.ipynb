# Iris Dataset exploration
## Creating some intuition about the data

```python
import pandas as pd
import seaborn as sbr
import matplotlib.pyplot as plt
import numpy as np
```

```python
#original dataset of UCI come without the column names
df = pd.read_csv("iris.data", header=-1)
columns = ['sepal length','sepal width','petal length','petal width','class']

#adding the real column names
df.columns = columns

#in case of your column names come with bad formats
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
# Visualization of the complete data
df.head()
```

```python
df.describe()
```

```python
# Comparing the results for sepal length mean and standart deviation
stdSL = np.std(df['sepal_length'])
meanSL = np.mean(df['sepal_length'])

print("Standart Deviation: ", stdSL,"\nMean: " ,meanSL,"\n")

# Seeing the species distribution 
print(df.groupby('class').size())
```

# Visualization

```python
#plt.figure(figsize=(10,8))
#subplot(2,2,1)
sbr.pairplot(df, hue='class', height=3);
```

```python
plt.figure(figsize=(12,9));
plt.subplot(2,2,1)
sbr.violinplot(x='class', y='sepal_length',scale='count',width =0.7, data=df)
plt.subplot(2,2,2)
sbr.violinplot(x='class',y='sepal_width',scale='count',width =0.7, data=df)
plt.subplot(2,2,3)
sbr.violinplot(x='class',y='petal_length',scale='count',width =0.7, data=df)
plt.subplot(2,2,4)
sbr.violinplot(x='class', y='petal_width',scale='count',width =0.7, data=df);
```
